* Of the 1163 active players, the vast majority are male (84%). There also exists, a smaller, but notable proportion of female players (14%).

* Our peak age demographic falls between 20-24 (44.8%) with secondary groups falling between 15-19 (18.60%) and 25-29 (13.4%).

```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# Raw data file
file_to_load = "Resources/purchase_data.csv"

# Read purchasing file and store into pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
# Total number of players
num_players = int(len(purchase_data.groupby("SN")))

# Display in DataFrame
pd.DataFrame([num_players], columns = ["Total Players"])
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# List of unique items
unq_items = pd.unique(purchase_data["Item Name"])

# Average Price
unq_price = pd.unique(purchase_data["Price"])
avg_price = unq_price.sum() / len(unq_items)

# Number of Purchases
purch = purchase_data.count()[1]

# Total Revenue
rev = purchase_data["Price"].sum()

# Summary DataFrame
summary = pd.DataFrame({"Number of Unique Items": len(unq_items),
                        "Average Price": [avg_price],
                        "Number of Purchases": purch,
                        "Total Revenue": rev})

# Format and display results
summary["Average Price"] = summary["Average Price"].map("${:.2f}".format)
summary["Total Revenue"] = summary["Total Revenue"].map("${:,.2f}".format)
summary
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# Make large frame of counts and prices by gender 
gender_group = purchase_data.groupby(["Gender", "SN"])
gender_count_obj = gender_group.count()
gender_count_obj = gender_count_obj.reset_index()
gender_sum_obj = gender_group.sum()
gender_sum_obj = gender_sum_obj.reset_index()
gender_obj = pd.merge(gender_sum_obj, gender_count_obj, on = ["SN", "Gender"], suffixes = [" Sum", " Count"])

# Count of usernames by genders
gender_count = gender_count_obj["Gender"].value_counts()

# Define percentages
perc1 = (gender_count[1] / num_players) * 100
perc2 = (gender_count[0] / num_players) * 100
perc3 = (gender_count[2] / num_players) * 100

# summary table
summary = pd.DataFrame({"Percentage of Players": [perc1, perc2, perc3],
                        "Total Count": [gender_count[1],
                                        gender_count[0],
                                        gender_count[2]]
                       }, index = ["Female", "Male", "Other / Non-Disclosed"])
summary["Percentage of Players"] = summary["Percentage of Players"].map("{:.2f}%".format)
summary
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Purchase count by gender
purchase_sum_f = gender_count_obj.loc[gender_count_obj["Gender"] == "Female", "Purchase ID"].sum()
purchase_sum_m = gender_count_obj.loc[gender_count_obj["Gender"] == "Male", "Purchase ID"].sum()
purchase_sum_o = gender_count_obj.loc[gender_count_obj["Gender"] == "Other / Non-Disclosed", "Purchase ID"].sum()

# Check total matches previously calculated total
purchase_sum_f + purchase_sum_m + purchase_sum_o

# Total purchase value by gender
total_spent_f = gender_sum_obj.loc[gender_sum_obj["Gender"] == "Female", "Price"].sum()
total_spent_m = gender_sum_obj.loc[gender_sum_obj["Gender"] == "Male", "Price"].sum()
total_spent_o = gender_sum_obj.loc[gender_sum_obj["Gender"] == "Other / Non-Disclosed", "Price"].sum()

# Check total matches previously calculated total
total_spent_f + total_spent_m + total_spent_o

# Average Purchase Price by gender
avg_price_f = total_spent_f / purchase_sum_f
avg_price_m = total_spent_m / purchase_sum_m
avg_price_o = total_spent_o / purchase_sum_o

# Average Purchase Price per person by gender
gender_obj["Avg Purchase Total"] = gender_obj["Price Sum"] / gender_obj["Price Count"]

# Average purchase total per person
avg_total_f = gender_sum_obj.loc[gender_sum_obj["Gender"] == "Female", "Price"].sum() / gender_count[1]
avg_total_m = gender_sum_obj.loc[gender_sum_obj["Gender"] == "Male", "Price"].sum() / gender_count[0]
avg_total_o = gender_sum_obj.loc[gender_sum_obj["Gender"] == "Other / Non-Disclosed", "Price"].sum() / gender_count[2]

# summary table
summary = pd.DataFrame({"Purchase Count": [purchase_sum_m, purchase_sum_f, purchase_sum_o],
                        "Average Purchase Price": [avg_price_m, avg_price_f, avg_price_o],
                        "Total Purchase Value": [total_spent_m, total_spent_f, total_spent_o],
                        "Avg Purchase Total per Person": [avg_total_m, avg_total_f, avg_total_o]
                       }, index = ["Male", "Female", "Other / Non-Disclosed"])
summary["Average Purchase Price"] = summary["Average Purchase Price"].map("${:.2f}".format)
summary["Total Purchase Value"] = summary["Total Purchase Value"].map("${:,.2f}".format)
summary["Avg Purchase Total per Person"] = summary["Avg Purchase Total per Person"].map("${:,.2f}".format)
summary
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Establish bins for ages
age_bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
purchase_data["Age Bin"] = pd.cut(purchase_data["Age"], age_bins, labels = group_names)

# Count and Sum tables 
age_group = purchase_data.groupby(["Age Bin", "SN"])
age_count_obj = pd.DataFrame(age_group.count())
age_count_obj = age_count_obj.reset_index()
age_count_obj = age_count_obj.dropna()

age_sum_obj = pd.DataFrame(age_group.sum())
age_sum_obj = age_sum_obj.reset_index()
age_sum_obj = age_sum_obj.dropna()

age_obj = pd.merge(age_sum_obj, age_count_obj, on = ["SN", "Age Bin"], suffixes = [" Sum", " Count"])

# Purchase count by age group
age_counts = []
for i in group_names:
    age_count = age_count_obj.loc[age_count_obj["Age Bin"] == i, "Purchase ID"].count()
    age_counts.append(age_count)

percent_age = [(x/num_players) * 100 for x in age_counts]

summary = pd.DataFrame({"Percentage of Players": percent_age,
                       "Total Count": age_counts},
                      index = group_names)
summary["Percentage of Players"] = summary["Percentage of Players"].map("{:.2f}%".format)
summary
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Total purchase price 
total_per_age = []
for i in group_names:
    total_count = age_count_obj.loc[age_count_obj["Age Bin"] == i, "Price"].sum()
    total_per_age.append(total_count)

# Total spent per age group    
total_spent_age = []
for i in group_names:
    total_sums = age_sum_obj.loc[age_sum_obj["Age Bin"] == i, "Price"].sum()
    total_spent_age.append(total_sums)

# Average Purchase price: total spent / total purchases 
avg_spent_age = [x/y for x, y in zip(total_spent_age,  total_per_age)]

# Average Purchase per person: total spent / number in age range
avg_per_person = [x/y for x, y in zip(total_spent_age,  age_counts)]
        
# summary table
summary = pd.DataFrame({"Purchase Count": total_per_age,
                       "Average Purchase Price": avg_spent_age,
                       "Total Purchase Value": total_spent_age,
                       "Avg Purchase Total per Person": avg_per_person},
                      index = group_names)
summary["Purchase Count"] = summary["Purchase Count"].map("{:.0f}".format)
summary["Average Purchase Price"] = summary["Average Purchase Price"].map("${:.2f}".format)
summary["Total Purchase Value"] = summary["Total Purchase Value"].map("${:,.2f}".format)
summary["Avg Purchase Total per Person"] = summary["Avg Purchase Total per Person"].map("${:,.2f}".format)
summary
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Get counts and sums of items
items = purchase_data.groupby(["Item ID", "Item Name"])
items_count = items.count()
items_sum = items.sum()
items = pd.merge(items_count, items_sum, on = ["Item ID", "Item Name"], suffixes = [" Count", " Sum"])

# Extract prices and add to the items frame
price = purchase_data[["Item ID", "Item Name", "Price"]]
price = pd.DataFrame(price.drop_duplicates())
items = pd.merge(items, price, on = ["Item ID", "Item Name"])
items = items[["Item ID", "Item Name", "Purchase ID Count", "Price", "Price Sum"]]

# Clean up the items data frame
items = items.set_index(["Item ID", "Item Name"])
items = items.sort_values("Purchase ID Count", ascending = False)
items = items.rename(columns = {"Purchase ID Count" : "Purchase Count",
                               "Price" : "Item Price",
                               "Price Sum" : "Total Purchase Value"})
items["Item Price"] = items["Item Price"].map("${:.2f}".format)
items["Total Purchase Value"] = items["Total Purchase Value"].map("${:.2f}".format)
items.head()
```
